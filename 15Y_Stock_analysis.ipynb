```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("15Year_stock.csv")

df.isnull().sum()
```

```python


stocks =["AAPL","AMZN","GOOGL","MSFT","NVDA"]

for stock in stocks :
 plt.figure(figsize=(12,6))
 date = df["Date"]
 Close_AAPL = df[f"Close_{stock}"]
 high_price_index = Close_AAPL.idxmax()
 print(f"The Highest Closing Price of {stock} is {Close_AAPL[high_price_index]} on {date[high_price_index]}")
 plt.plot(date,Close_AAPL,linestyle="-",color="red",label=f"{stock} Closing Price")
 plt.xlabel("Date")
 plt.ylabel("Closing Price")
 plt.grid(True)
 plt.xticks(ticks=range(0,len(date),120),labels=date[::120],rotation=45)
 plt.legend()
 plt.title(f"Change in Closing price over every 6 months of {stock}")
 plt.tight_layout()
 plt.savefig(f"{stock}_Closing_Price_over_every_6_Months.png")
 plt.show()
```
