```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("15Year_stock.csv")

df.isnull().sum()
```

1. How did the closing prices of each stock change over time?

```python


stocks =["AAPL","AMZN","GOOGL","MSFT","NVDA"]
plt.figure(figsize=(14,8))
for stock in stocks :
    
    date = df["Date"]
    Close_AAPL = df[f"Close_{stock}"]
    high_price_index = Close_AAPL.idxmax()
    print(f"The Highest Closing Price of {stock} is {Close_AAPL[high_price_index]} on {date[high_price_index]}")
    plt.plot(date,Close_AAPL,linestyle="-",label=f"{stock} Closing Price")
    
plt.xlabel("Date")
plt.ylabel("Closing Price")
plt.grid(True)
plt.xticks(ticks=range(0,len(date),120),labels=date[::120],rotation=45)
plt.legend()
plt.title(f"Compariosn of change in closing price")
plt.tight_layout()
plt.savefig(f"Compariosn_of_Closing_Price_over_every_3_Months.png")
plt.show()
```

2. Which company had the highest trading volume over time?

```python

stocks =["AAPL","AMZN","GOOGL","MSFT","NVDA"]



for stock in stocks :
 plt.figure(figsize=(12,6))
 date = df["Date"]
 Volume_AAPL = df[f"Volume_{stock}"]
 high_price_index = Volume_AAPL.idxmax()
 print(f"The Highest Volume Price of {stock} is {Volume_AAPL[high_price_index]} on {date[high_price_index]}")
 plt.plot(date,Volume_AAPL,linestyle="-",color="teal",label=f"{stock} Volume Price")
 plt.xlabel("Date")
 plt.ylabel("Volume Price")
 plt.grid(True)
 plt.xticks(ticks=range(0,len(date),120),labels=date[::120],rotation=45)
 plt.legend()
 plt.title(f"Change in Volume price over every 3 months of {stock}")
 plt.tight_layout()
 plt.savefig(f"{stock}_Volume_Price_over_every_3_Months.png")
 plt.show()
```

3.  Which stock had the widest average daily price range (High - Low) over the entire period?

```python
stocks = ["AAPL","AMZN","GOOGL","MSFT","NVDA"]

avg_daily_price =[]

for stock in stocks :
    high_value = df[f"High_{stock}"]
    low_value = df[f"Low_{stock}"]
    avg_value = (high_value-low_value).mean()
    avg_daily_price.append(avg_value)

plt.figure(figsize=(12,6))
plt.bar(stocks,avg_daily_price,color="red")
plt.title("Average Daily price range over entire Period")
plt.xlabel("Stock Name")
plt.ylabel("Range")
plt.legend()
plt.tight_layout()
plt.grid(True)
plt.savefig("avg_daily_price.png",dpi=300)
plt.show()
```

4. Which stock had the maximum single-day price range, and when?


```python
stocks = ["AAPL","AMZN","GOOGL","MSFT","NVDA"]

avg_daily_price =[]
max_range_value =0
max_value_name =""
max_value_idx =0
date = df["Date"]

for stock in stocks :
    high_value = df[f"High_{stock}"]
    low_value = df[f"Low_{stock}"]
    avg_value = (high_value - low_value).max()
    max_value_idxx = (high_value - low_value).idxmax()
    if(max_range_value < avg_value):
        max_range_value = avg_value
        max_value_name = stock
        max_value_idx = max_value_idxx

    avg_daily_price.append(avg_value)

plt.figure(figsize=(12,6))
plt.bar(stocks,avg_daily_price,color="red",label="Sinle day Price range")
plt.title("Single Day Price Range Comparison")
plt.xlabel("Stock Name")
plt.ylabel("Price Range")
plt.legend()
plt.tight_layout()
plt.grid(True)
plt.savefig("single_day_price.png",dpi=300)
plt.show()

print(f"The {max_value_name} has maximum single-day price range on {date[max_value_idx]} and the Price is {max_range_value}")
```

5. Which stock showed the highest volatility over the time period? ( Calculate Standard deviation )

```python
stocks = ["AAPL","AMZN","GOOGL","MSFT","NVDA"]

std_value =[]

for stock in stocks :
    closing_price = df[f"Close_{stock}"]
    std = closing_price.std()
    std_value.append(std)
  
plt.figure(figsize=(12,6))
plt.bar(stocks,std_value,color="skyblue",label="Volatility over time")
plt.title("Volatility of Each stock over time")
plt.xlabel("Stocks")
plt.ylabel("Volatility Range")
plt.ylim(0,150)
plt.grid(True)
plt.tight_layout()
plt.savefig("highest_volatility_over_time.png")
plt.show()
```
